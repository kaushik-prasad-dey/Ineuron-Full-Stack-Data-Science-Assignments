Q1. Explain what inheritance is in object-oriented programming and why it is used?

As a fundamental concept in software development, inheritance enables the creation of new classes that inherit attributes and behaviours from existing classes,improving code reusability and overall efficiency.

In object-oriented programming (OOP), inheritance is a framework that allows one class (the child or subclass) to inherit the properties and behaviors (attributes and methods) of another class (the parent or superclass).

This means that a subclass can reuse and extend the features of its superclass, promoting code reusability, and establishing a hierarchical relationship between classes.

Here are some key concepts related to inheritance:

**Base Class (Superclass):** The class whose attributes and methods are inherited by another class. It's also known as the superclass or parent class.

**Derived Class (Subclass):** The class that inherits attributes and methods from another class. It's also known as the subclass or child class.

**"is-a" Relationship:** Inheritance is often used to model an "is-a" relationship. For example, if you have a Vehicle class, you might have subclasses like Car and Motorcycle, as both are types of vehicles.

```python
#one simple code example of inheritence
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}")

class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type

    def display_info(self):
        super().display_info()
        print(f"Fuel Type: {self.fuel_type}")

if __name__=="__main__":
   vehicle = Vehicle(make="Toyota", model="Camry", year=2022)
   car = Car(make="Tesla", model="Model 3", year=2023, fuel_type="Electric")
   # Accessing methods of both Vehicle and Car
   vehicle.display_info()
   print("\n")
   car.display_info()
   print( vehicle.__init__ )
```

In object-oriented programming, inheritance can be classified into two main types:

`single` and `multiple` inheritance.

Single Inheritance:

Definition: Single inheritance occurs when a class inherits from only one base class. The subclass inherits the attributes and behaviors of a single parent class.

**Example of single inheritence**

```python
#here base class is animal
class Animal:
    def speak(self):
        print("Animal speaks")

#subclass Dog is inherited from the base class Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

if __name__=="__main__":
  dog = Dog()
  dog.speak()  # Inherited from Animal
  dog.bark()   # Defined in Dog
```

```python
#Another example of single inheritence
class Parent:
  def f1(self):
    print("Function of parent class.")

class Child(Parent):
  def f2(self):
    print("Function of child class.")

object1 = Child()
object1.f1()
object1.f2()
```

**The advantages of Single inheritence**

Simplicity: Single inheritance is straightforward and easy to understand.

Reduced Complexity: There's a clear and linear hierarchy of classes.

**Multiple Inheritance :**

Definition: Multiple inheritance occurs when a class inherits from more than one base class.
The subclass inherits the attributes and behaviors of multiple parent classes.

**Example of multiple inheritence**

```python
class Flyable:
    def fly(self):
        print("Can fly")

class Swimmable:
    def swim(self):
        print("Can swim")

#this Flyingfish can inheritence from Flyable & Swimmmable
class FlyingFish(Flyable, Swimmable):
    pass

if __name__=="__main__":
  flying_fish = FlyingFish()
  flying_fish.fly()   # Inherited from Flyable
  flying_fish.swim()  # Inherited from Swimmable
```

```python
#example of multiple inheritence
class Parent_1:
  def f1(self):
    print("Function of parent_1 class.")

class Parent_2:
  def f2(self):
    print("Function of parent_2 class.")

class Parent_3:
  def f3(self):
    print("function of parent_3 class.")

class Child(Parent_1, Parent_2, Parent_3):
  def f4(self):
    print("Function of child class.")

object_1 = Child()
object_1.f1()
object_1.f2()
object_1.f3()
object_1.f4()
```

**The advantages of multiple inheritence are given below:**

Enhanced Reusability: Allows a class to inherit from multiple sources, promoting code reuse.

Versatility: Enables the creation of classes with a combination of functionalities from different sources.

**Differences**

**Number of Base Classes:** The main difference is the number of base classes a subclass can inherit from.

**Hierarchy:** Single inheritance creates a linear hierarchy, while multiple inheritance can create a more complex hierarchy.

**Considerations:**

Diamond Problem: In multiple inheritance, the diamond problem can occur when a class inherits from two classes that have a common ancestor. It may lead to ambiguity in method resolution. In Python, the **method resolution order (MRO)** helps to determine the order in which base classes are searched.

Example of diamond problem

```python
class A:
    def speak(self):
        print("A speaks")

class B(A):
    def speak(self):
        print("B speaks")

class C(A):
    def speak(self):
        print("C speaks")

class D(B, C):
    pass

if __name__=="__main__":
   obj = D()
   obj.speak()  # Ambiguity arises in the presence of the diamond problem
```

Q3. Explain the terms `base class` and `derived class` in the context of inheritance ??

In the context of inheritance, the terms `base class` and `derived class` refer to the relationship between two classes, where the derived class inherits attributes and behaviors from the base class.

image.png

**Base Class (Superclass):** The class whose attributes and methods are inherited by another class. It's also known as the superclass or parent class.

**Derived Class (Subclass):** The class that inherits attributes and methods from another class. It's also known as the subclass or child class.

```python
#Python code example to illustrate the concepts of base class and derived class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

if __name__=="__main__":
  dog = Dog(name="Buddy")
  cat = Cat(name="Whiskers")
  # Accessing attributes and methods from the base class (Animal)
  print(f"{dog.name} says:")
  dog.speak()  # Inherited from Animal
  print("\n")
  print(f"{cat.name} says:")
  cat.speak()  # Inherited from Animal
  print("\n")
  # Accessing methods from the derived classes
  dog.bark()  # Defined in Dog
  cat.meow()  # Defined in Cat
```

**Explanation:**

Animal is the base class with an `__init__` method (constructor) and a speak method.

Dog and Cat are derived classes that inherit from the Animal class.

Both Dog and Cat classes have their own methods (bark and meow, respectively) in addition to inheriting the speak method from the Animal class.

When you create instances of the Dog and Cat classes, they have access to both their own methods and the methods of the base class (Animal).

This illustrates the concept of inheritance, where the derived class builds upon the functionality of the base class.

**Another 2 example of base class and derived class here**

```python
class Shape:
    def __init__(self, color):
        self.color = color

    def draw(self):
        print(f"Drawing a shape in {self.color}")

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def draw(self):  # Overriding the draw method in the derived class
        print(f"Drawing a circle in {self.color} with radius {self.radius}")

class Square(Shape):
    def __init__(self, color, side_length):
        super().__init__(color)
        self.side_length = side_length

    def draw(self):  # Overriding the draw method in the derived class
        print(f"Drawing a square in {self.color} with side length {self.side_length}")

if __name__=="__main__":
  circle = Circle(color="Red", radius=5)
  square = Square(color="Blue", side_length=4)
  circle.draw()  # Calls the draw method of the Circle class
  square.draw()  # Calls the draw method of the Square class
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print(f"Name: {self.name}, Salary: {self.salary}")

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def display_info(self):  # Overriding the display_info method in the derived class
        super().display_info()
        print(f"Department: {self.department}")

class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language

    def display_info(self):  # Overriding the display_info method in the derived class
        super().display_info()
        print(f"Programming Language: {self.programming_language}")

if __name__=="__main__":
   manager = Manager(name="Alice", salary=80000, department="IT")
   developer = Developer(name="Bob", salary=60000, programming_language="Python")
   manager.display_info()    # Calls the display_info method of the Manager class
   developer.display_info()  # Calls the display_info method of the Developer class
```

Q4. What is the significance of the `protected` access modifier in inheritance?
How does it differ from "private" and "public" modifiers?

In object-oriented programming, access modifiers like "public," "private," and "protected" control the visibility and accessibility of class members (attributes and methods).

The "protected" access modifier is significant in the context of inheritance as it allows members to be accessed by the derived (subclass) classes, but not by external classes or code.


Let's explore the significance of the "protected" modifier by comparing it with "private" and "public" modifiers:

image.png

**1. Public Access Modifier:**

**Visibility:** Public members are accessible from anywhere, both within the class and from external classes.



```python
#Example of public access Modifier
class BaseClass:
    def __init__(self):
        self.public_attribute = "I am public"

class DerivedClass(BaseClass):
    def display_info(self):
        print(self.public_attribute)

if __name__=="__main__":
  obj = DerivedClass()
  obj.display_info()  # Accessing the public attribute from the derived class
```

**2. Private Access Modifier:**

**Visibility:** Private members are only accessible within the class in which they are defined. They are not directly accessible in derived classes.

```python
class BaseClass:
    def __init__(self):
        self.__private_attribute = "I am private"

class DerivedClass(BaseClass):
    def display_info(self):
        # Attempting to access the private attribute using name mangling
        print(self._BaseClass__private_attribute)

if __name__=="__main__":
  obj = DerivedClass()
  obj.display_info()  # Raises an AttributeError due to name mangling

```

**3. Protected Access Modifier:**

Visibility: Protected members are accessible within the class in which they are defined and also in derived classes.

```python
class BaseClass:
    def __init__(self):
        self._protected_attribute = "I am protected"

class DerivedClass(BaseClass):
    def display_info(self):
        print(self._protected_attribute)

if __name__=="__main__":
  obj = DerivedClass()
  obj.display_info()  # Accessing the protected attribute from the derived class
```

so, we can say that

**Public:** Accessible everywhere.

**Private:** Accessible only within the class.

**Protected:** Accessible within the class and in derived classes.


While Python does not enforce strict access control like some other languages, using naming conventions (e.g., a single leading underscore for protected members) helps convey the intended visibility.

It's important to note that users of the class should respect these conventions to maintain the expected behavior and encapsulation.

Q5. What is the purpose of the `super` keyword in inheritance? Provide an example

Inheritance is a key concept in object-oriented programming (OOP) where a class (subclass or derived class) can inherit attributes and methods from another class (superclass or base class).

The `super` keyword plays a crucial role in inheritance by providing a way to call methods and constructors from the superclass within the subclass.

image.png

Here are the main purposes of the `super` keyword in inheritance:

**Accessing Superclass Methods:**

You can use super to call a method from the superclass, allowing you to reuse or override the behavior of the superclass method in the subclass.

**Accessing Superclass Constructors:**

When creating an instance of the subclass, the super keyword is used to call the constructor of the superclass.

This ensures that the initialization code in the superclass is executed before the initialization code in the subclass.

```python
#Python to illustrate the use of super in inheritance:
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name, breed):
        # Calling the constructor of the superclass (Animal)
        super().__init__(name)
        self.breed = breed

    def speak(self):
        # Calling the speak method of the superclass (Animal)
        super().speak()
        print(f"{self.name} barks")

# Example usage:
dog = Dog(name="Buddy", breed="Labrador")
dog.speak()
```

Q6. Create a base class called `Vehicle` with attributes like `make`, `model`, and `year`.
Then, create a derived class called `Car` that inherits from `Vehicle` and adds an
attribute called `fuel_type`.
Implement appropriate methods in both classes.


```python
#base class called Vehicle
class Vehicle:
    #first load the init constructor
    def __init__(self, make, model, year):
        #first attribute is make
        self.make = make
        #second attribute is model
        self.model = model
        #third attribute is year
        self.year = year

    def display_info(self):
        headers = ["Attribute", "Value"]
        data = [
            ["Make", self.make],
            ["Model", self.model],
            ["Year", self.year]
        ]
        print(f"Vehicle Information - {type(self).__name__}")
        print(tabulate(data, headers, tablefmt="grid"))

#create a derived class called Car that inherits from Vehicle
class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type

    def display_info(self):
        super().display_info()  # Call the display_info method of the base class
        print(f"Fuel Type: {self.fuel_type}")

if __name__=="__main__":
   vehicle = Vehicle(make="Toyota", model="Camry", year=2022)
   car = Car(make="Tesla", model="Model 3", year=2023, fuel_type="Electric")
   # Displaying information for both Vehicle and Car in tabular format
   vehicle.display_info()
   print("\n")
   car.display_info()
```

Q8. Design a base class called `Shape` with attributes like `colour` and `border_width.`
Create derived classes, `Rectangle` and `Circle,` that inherit from `Shape` and add specific attributes like `length` and `width` for the `Rectangle` class and `radius` for the `Circle` class.

```python
#create a base class shape
class Shape:
    def __init__(self, colour, border_width):
        self.colour = colour
        self.border_width = border_width

#now create the derived class Rectangle and call the base class Shape
class Rectangle(Shape):
    def __init__(self, colour, border_width, length, width):
        super().__init__(colour, border_width)
        self.length = length
        self.width = width

#now create the derived class Circle and call the base class Shape
class Circle(Shape):
    def __init__(self, colour, border_width, radius):
        super().__init__(colour, border_width)
        self.radius = radius

#now create the display shape infor and display the output with this class informations
def display_shapes_info(shapes):
    headers = ["Shape", "Colour", "Border Width", "Additional Info"]
    data = []

    for shape in shapes:
        if isinstance(shape, Rectangle):
            data.append(["Rectangle", shape.colour, shape.border_width, f"Length: {shape.length}, Width: {shape.width}"])
        elif isinstance(shape, Circle):
            data.append(["Circle", shape.colour, shape.border_width, f"Radius: {shape.radius}"])

    print(tabulate(data, headers, tablefmt="grid"))

#entry point class which is bootstrapping first when code runs
if __name__=="__main__":
  rectangle = Rectangle(colour="Red", border_width=2, length=5, width=3)
  circle = Circle(colour="Blue", border_width=1, radius=4)
  shapes = [rectangle, circle]
  # Displaying shape information in a tabular format
  display_shapes_info(shapes)
```

Q9. Create a base class called 'Device' with attributes like `brand` and 'model.'
Derive two classes, `Phone` and `Tablet,` from `Device.` Add specific attributes like
`screen_size` for the `Phone` class and `battery_capacity` for the `Tablet` class.


```python
#create the base class Device
class Device:
    def __init__(self, brand, model):
        #first attribute is brand
        self.brand = brand
        #second attribute is model
        self.model = model

#create the class Phone and derived from Device
class Phone(Device):
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size

#create the class Tablet and derived from Device
class Tablet(Device):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity

def display_devices_info(devices):
    headers = ["Device Type", "Brand", "Model", "Additional Info"]
    data = []

    for device in devices:
        if isinstance(device, Phone):
            data.append(["Phone", device.brand, device.model, f"Screen Size: {device.screen_size} inches"])
        elif isinstance(device, Tablet):
            data.append(["Tablet", device.brand, device.model, f"Battery Capacity: {device.battery_capacity} mAh"])

    print(tabulate(data, headers, tablefmt="grid"))

#entry point class which is bootstrapping first when code runs
if __name__=="__main__":
  iphone = Phone(brand="Apple", model="iPhone 12", screen_size=6.1)
  ipad = Tablet(brand="Apple", model="iPad Pro", battery_capacity=9720)
  devices = [iphone, ipad]
  # Displaying device information in a tabular format
  display_devices_info(devices)
```

Q10. Create a base class called `BankAccount` with attributes like `account_number` and `balance.`
Derive two classes, `SavingsAccount` and `CheckingAccount,` from
`BankAccount.` Add specific methods like `calculate_interest` for the
`SavingsAccount` class and `deduct_fees` for the `CheckingAccount` class.

```python
#base class called BankAccount
class BankAccount:
    #first load the init constructor
    def __init__(self, account_number, balance):
        #first attribute account number
        self.account_number = account_number
        #second attribute balance
        self.balance = balance
#now derived class savingsAccount with inheritence from base class
class SavingsAccount(BankAccount):
    #first load the init constructor
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate
    #create udf for calculate interest
    def calculate_interest(self):
        interest = self.balance * (self.interest_rate / 100)
        self.balance += interest
        return interest
#now derived class checkingAccount with inheritence from base class
class CheckingAccount(BankAccount):
     #first load the init constructor
    def __init__(self, account_number, balance, fee_percentage):
        super().__init__(account_number, balance)
        self.fee_percentage = fee_percentage
    #create udf for deduct fees
    def deduct_fees(self):
        fees = self.balance * (self.fee_percentage / 100)
        self.balance -= fees
        return fees

#create display account info for showing data in tabular format
def display_account_info(accounts):
    headers = ["Account Type", "Account Number", "Balance", "Additional Info"]
    data = []

    for account in accounts:
        if isinstance(account, SavingsAccount):
            data.append(["Savings", account.account_number, account.balance, f"Interest Rate: {account.interest_rate}%"])
        elif isinstance(account, CheckingAccount):
            data.append(["Checking", account.account_number, account.balance, f"Fee Percentage: {account.fee_percentage}%"])

    print(tabulate(data, headers, tablefmt="grid"))

#entry point class which is bootstrapping first when code runs
if __name__=="__main__":
  savings_account = SavingsAccount(account_number="S123456", balance=1000, interest_rate=2.5)
  checking_account = CheckingAccount(account_number="C789012", balance=1500, fee_percentage=1.0)
  accounts = [savings_account, checking_account]
  # Perform specific actions on each account
  interest_earned = savings_account.calculate_interest()
  fees_deducted = checking_account.deduct_fees()
  # Displaying account information in a tabular format
  display_account_info(accounts)
  # Displaying additional details after performing actions
  print("\nInterest earned on Savings Account:", interest_earned)
  print("Fees deducted from Checking Account:", fees_deducted)
```
