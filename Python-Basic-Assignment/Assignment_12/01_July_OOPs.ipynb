 Q1. **What is the primary goal of Object-Oriented Programming (OOP)?**

image.png

The primary goal of Object-Oriented Programming (OOP) is to design software in a way that models real-world entities and their interactions. OOP is centered around the concept of 'objects' which are instances of classes.

A class is a blueprint or a template that defines the properties and behaviors common to all objects of that type.



The four main principles of OOP are:

image.png

**Encapsulation**: This involves bundling the data (attributes) and the methods (functions) that operate on the data into a single unit called a class. Encapsulation helps hide the internal details of how an object works and exposes only what is necessary.

**Inheritance:** This allows a new class to inherit the characteristics and behaviors of an existing class. It promotes code reuse and the creation of a hierarchy of classes, where a subclass (derived class) can inherit attributes and methods from a superclass (base class).

**Polymorphism:** This refers to the ability of objects of different classes to be treated as objects of a common base class. Polymorphism allows for flexibility and extensibility in code by enabling objects to be used interchangeably.

**Abstraction:** Abstraction involves simplifying complex systems by modeling classes based on the essential properties and behaviors they share. It allows developers to focus on what an object does rather than how it achieves its functionality.

**Coding example of Encapsulation**



````python
class Car:
    def __init__(self, make, model):
        self.__make = make  # private attribute
        self.__model = model  # private attribute
        self.__fuel = 100  # private attribute, initial fuel level

    def get_make(self):
        return self.__make

    def get_model(self):
        return self.__model

    def get_fuel(self):
        return self.__fuel

    def drive(self, distance):
        fuel_needed = distance / 10  # assuming 10 km per liter
        if fuel_needed <= self.__fuel:
            print(f"Driving {distance} km.")
            self.__fuel -= fuel_needed
        else:
            print("Not enough fuel!")

# Creating an instance of the Car class
my_car = Car("Toyota", "Camry")

# Accessing attributes using getter methods
print(f"Make: {my_car.get_make()}")
print(f"Model: {my_car.get_model()}")
print(f"Fuel level: {my_car.get_fuel()}")

# Trying to access private attribute directly (will result in an error)
# print(my_car.__make)  # This line will raise an AttributeError

# Driving the car
my_car.drive(50)

# Checking the fuel level again
print(f"Fuel level: {my_car.get_fuel()}")
````

**Coding example of Inheritance**

````python
# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # Placeholder for the speak method

# Derived class inheriting from Animal
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Another derived class inheriting from Animal
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

# Creating instances of the derived classes
dog_instance = Dog("Buddy")
cat_instance = Cat("Whiskers")

# Using the speak method of the derived classes
print(dog_instance.speak())  # Output: Buddy says Woof!
print(cat_instance.speak())  # Output: Whiskers says Meow!

# Accessing the attribute from the base class
print(f"{dog_instance.name} is a friendly dog.")  # Output: Buddy is a friendly dog.
````

**Coding example of Polymorphism**



````python
class Bird:
    def sound(self):
        pass  # Placeholder for the sound method

class Sparrow(Bird):
    def sound(self):
        return "Chirp!"

class Parrot(Bird):
    def sound(self):
        return "Squawk!"

# Function that takes an object of the Bird class and calls its sound method
def make_sound(bird):
    return bird.sound()

# Creating instances of the derived classes
sparrow_instance = Sparrow()
parrot_instance = Parrot()

# Using the make_sound function with different types of birds
print(make_sound(sparrow_instance))  # Output: Chirp!
print(make_sound(parrot_instance))   # Output: Squawk!
````

**Coding example of Abstraction**

````python
from abc import ABC, abstractmethod

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Concrete class implementing Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Concrete class implementing Shape
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

    def perimeter(self):
        return 4 * self.side

# Function that takes any object of type Shape and prints its area and perimeter
def print_shape_info(shape):
    print(f"Area: {shape.area()}")
    print(f"Perimeter: {shape.perimeter()}")

# Creating instances of the derived classes
circle_instance = Circle(5)
square_instance = Square(4)

# Using the print_shape_info function with different types of shapes
print("Circle:")
print_shape_info(circle_instance)

print("\nSquare:")
print_shape_info(square_instance)
````

Q2. **What is an object in Python?**

image.png

In Python, an object is a fundamental concept that represents a piece of data along with the methods or functions that operate on that data. Everything in Python is an object, and every object has a type (class) that defines the characteristics (attributes) and behaviors (methods) associated with it.

Here are some key points about objects in Python:

1. **Instance of a Class**: An object is an instance of a class. A class is a blueprint or template that defines the attributes and methods common to all objects of that type.

2. **Attributes:** Objects have attributes, which are variables that store data. These attributes are defined in the class.

3. **Methods:** Objects have methods, which are functions that perform actions or operations on the data stored in the object. These methods are also defined in the class.

4. **Dynamic Typing:** Python is a dynamically-typed language, which means you don't need to explicitly declare the type of an object. The type is determined at runtime.

5. **Object Identity:** Each object in Python has a unique identity, which can be obtained using the id() function. Two objects with the same value can have different identities.

6. **Object-oriented Paradigm:** Python is an object-oriented programming (OOP) language, which means it emphasizes the use of objects and classes to structure code.

**Coding example of objects in python**

````python
# Define a simple class
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print("Woof!")

# Create instances (objects) of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Access attributes and call methods of the objects
print(f"{dog1.name} is {dog1.age} years old.")
dog1.bark()

print(f"{dog2.name} is {dog2.age} years old.")
dog2.bark()
````

Q4. **What are attributes and methods in a class?**

image.png

In a Python class, attributes and methods are fundamental components that define the structure and behavior of objects created from that class.

**Attributes:**

Attributes are variables that store data associated with an object.
They represent the state or characteristics of an object.
Attributes are defined within the class and are accessed using the dot notation with the instance of the class.
They are typically initialized in the class constructor (__init__ method).
Attributes can be public (accessible from outside the class), protected (accessible within the class and its subclasses), or private (accessible only within the class).

**Example of attributes**

````python
class Dog:
    def __init__(self, name, age):
        self.name = name  # Public attribute
        self._age = age    # Protected attribute
        self.__weight = 20  # Private attribute
````

````python
my_dog = Dog("Buddy", 3)
print(my_dog.name)       # Accessing public attribute
print(my_dog._age)        # Accessing protected attribute (convention, not enforced)
# print(my_dog.__weight)  # This line would raise an AttributeError (private attribute)
````

**Methods:**

Methods are functions defined within a class that perform actions or operations on the object's data (attributes).
They represent the behavior of the object.
Methods can take parameters, including the special parameter self, which refers to the instance of the object calling the method.
They are called using the dot notation on an instance of the class.
Methods can be public, protected, or private, similar to attributes.

**Example of Methods**

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):  # Public method
        print("Woof!")

    def _sleep(self):  # Protected method
        print(f"{self.name} is sleeping.")

    def __eat(self):  # Private method
        print(f"{self.name} is eating.")
````

````python
my_dog = Dog("Buddy", 3)
my_dog.bark()      # Calling public method
my_dog._sleep()    # Calling protected method (convention, not enforced)
# my_dog.__eat()   # This line would raise an AttributeError (private method)
````

Attributes and methods together define the structure and behavior of objects created from a class, providing a way to model real-world entities in a program.

**Example of Class Variables:**

````python
class Dog:
    # Class variable
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

# Accessing class variable using the class name
print(Dog.species)  # Output: Canis familiaris

# Creating instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Accessing instance variables
print(dog1.name, dog1.age)  # Output: Buddy 3
print(dog2.name, dog2.age)  # Output: Max 5
````

**Example of Instance Variables**

````python
class Car:
    def __init__(self, make, model, year):
        # Instance variables
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"{self.year} {self.make} {self.model}")

# Creating instances of the Car class
car1 = Car("Toyota", "Camry", 2022)
car2 = Car("Honda", "Accord", 2021)

# Accessing instance variables and calling a method
print(car1.make, car1.model, car1.year)  # Output: Toyota Camry 2022
car1.display_info()  # Output: 2022 Toyota Camry
print(car2.make, car2.model, car2.year)  # Output: Honda Accord 2021
car2.display_info()  # Output: 2021 Honda Accord
````

Q6. **What is the purpose of the self parameter in Python class methods?**

image.png


In Python, the self parameter in class methods is a convention that refers to the instance of the class itself. It is the first parameter of any method in a class and is used to access and manipulate the attributes and methods of the instance. The use of self is a standard practice in Python, although you could technically use any name for this parameter (but self is widely adopted and recommended).

Here are the main purposes of the `self` parameter in Python class methods:

**Instance Reference:**

self is used to refer to the instance of the class. It allows methods to access and modify the attributes of the specific instance on which the method is called.

**Attribute Access:**

Inside a class method, self is used to access instance variables (attributes) and other methods. It distinguishes instance variables from local variables within the method

**Method Invocation:**

self is required to call other methods of the same class. This ensures that the correct method is called on the specific instance.

**Instance Creation:**

When an instance of a class is created, the instance is automatically passed as the first argument to the __init__ method, which initializes the instance attributes. self allows you to refer to and initialize the attributes of the newly created instance.

Here's a simple example to illustrate the use of self in a class:



````python
class Dog:
    def __init__(self, name, age):
        # Initialize instance variables
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says Woof!")

# Creating an instance of the Dog class
my_dog = Dog("Buddy", 3)

# Accessing instance variables using self
print(f"{my_dog.name} is {my_dog.age} years old.")

# Calling a method using self
my_dog.bark()  # Output: Buddy says Woof!
````

In this example, self is used in the `__init__` method to initialize the name and age attributes of the instance. It is also used in the bark method to access the name attribute of the specific instance and print a message. The use of self helps differentiate instance variables from local variables and allows for proper instance-specific behavior.

Q8. For a ticket booking system, you have to design the 'Ticket' class with OOP
principles in mind.

The `Ticket` class should have the following attributes:
a. ticket_id: Represents the unique identifier for the ticket.

b. event_name: Represents the name of the event.

c. event_date: Represents the date of the event.

d. venue: Represents the venue of the event.

e. seat_number: Represents the seat number associated with the ticket.

f. price: Represents the price of the ticket.

g. is_reserved: Represents the reservation status of the ticket.

**The class also includes the following methods:**

a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.

b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
reserved.

c. display_ticket_info(self): Displays the information about the ticket, including its
attributes and reservation status.


**Install JS Animation libraries**

````python
!pip install JSAnimation
````

````python
from tabulate import tabulate

class Ticket:
    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price, is_reserved=False):
        # Attributes
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved

    def reserve_ticket(self):
        if not self.is_reserved:
            self.is_reserved = True
            print(f"Ticket {self.ticket_id} reserved successfully.")
        else:
            print(f"Ticket {self.ticket_id} is already reserved.")

    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False
            print(f"Reservation for Ticket {self.ticket_id} canceled.")
        else:
            print(f"Ticket {self.ticket_id} is not reserved.")

    def display_ticket_info(self):
        ticket_info = {
            "Ticket ID": self.ticket_id,
            "Event Name": self.event_name,
            "Event Date": self.event_date,
            "Venue": self.venue,
            "Seat Number": self.seat_number,
            "Price": f"${self.price:.2f}",
            "Is Reserved": 'Yes' if self.is_reserved else 'No'
        }

        # Display ticket information as a table
        print(tabulate([ticket_info], headers="keys", tablefmt="grid"))

# Create an instance of the Ticket class
ticket = Ticket(ticket_id=1, event_name="Concert", event_date="2023-01-01", venue="Arena", seat_number="A101", price=50.0)

# Function to update the plot during the animation
def update_plot(frame):
    # Displaying ticket information
    ticket.display_ticket_info()
    # Reserving the ticket
    ticket.reserve_ticket()
    # Displaying updated ticket information
    ticket.display_ticket_info()
    # Canceling the reservation
    ticket.cancel_reservation()
    # Displaying final ticket information
    ticket.display_ticket_info()

    # Clear the previous plot and display the updated book information
    plt.clf()
    plt.text(0.5, 0.5, f"Reservation Status: {ticket.is_reserved}", ha='center', va='center', fontsize=12)

# Create a figure and axis
fig, ax = plt.subplots()
# Create an animation
ani = animation.FuncAnimation(fig, update_plot, frames=5, interval=2000, repeat=False)
# Display the animation using HTML and JSAnimation
HTML(ani.to_jshtml())
````

Q9. You are creating a shopping cart for an e-commerce website. Using OOP to model

the `ShoppingCart` functionality the class should contain following attributes and methods:

a. items: Represents the list of items in the shopping cart.
The class also includes the following methods:

a. add_item(self, item): Adds an item to the shopping cart by appending it to the
list of items.

b. remove_item(self, item): Removes an item from the shopping cart if it exists in the list.

c. view_cart(self): Displays the items currently present in the shopping cart.

d. clear_cart(self): Clears all items from the shopping cart by reassigning an empty list to the items attribute.

````python
class ShoppingCart:
    def __init__(self):
        # Attribute
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"Item '{item}' added to the shopping cart.")

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print(f"Item '{item}' removed from the shopping cart.")
        else:
            print(f"Item '{item}' not found in the shopping cart.")

    def view_cart(self):
        if not self.items:
            print("Shopping cart is empty.")
        else:
            print("Items in the shopping cart:")
            for item in self.items:
                print(f"- {item}")

    def clear_cart(self):
        self.items = []
        print("Shopping cart cleared.")

# Example usage:

# Creating an instance of the ShoppingCart class
cart = ShoppingCart()

# Adding items to the shopping cart
cart.add_item("Laptop")
cart.add_item("Headphones")
cart.add_item("Mouse")

# Viewing the current shopping cart
cart.view_cart()

# Removing an item from the shopping cart
cart.remove_item("Headphones")

# Viewing the updated shopping cart
cart.view_cart()

# Clearing the shopping cart
cart.clear_cart()

# Viewing the empty shopping cart
cart.view_cart()
````

````python
class ShoppingCart:
    def __init__(self):
        # Attribute
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        return f"Item '{item}' added to the shopping cart."

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            return f"Item '{item}' removed from the shopping cart."
        else:
            return f"Item '{item}' not found in the shopping cart."

    def view_cart(self):
        if not self.items:
            return "Shopping cart is empty."
        else:
            cart_content = "Items in the shopping cart:"
            for item in self.items:
                cart_content += f"\n- {item}"
            return cart_content

    def clear_cart(self):
        self.items = []
        return "Shopping cart cleared."

# Streamlit app
def shopping_cart_app():
    st.title("Shopping Cart App")

    # Create a session-specific ShoppingCart instance
    if 'cart' not in st.session_state:
        st.session_state.cart = ShoppingCart()

    option = st.selectbox("Choose an action:", ["Add Item", "Remove Item", "View Cart", "Clear Cart"])

    if option == "Add Item":
        item_to_add = st.text_input("Enter item to add:")
        if st.button("Add"):
            result = st.session_state.cart.add_item(item_to_add)
            st.success(result)

    elif option == "Remove Item":
        item_to_remove = st.text_input("Enter item to remove:")
        if st.button("Remove"):
            result = st.session_state.cart.remove_item(item_to_remove)
            st.success(result)

    elif option == "View Cart":
        result = st.session_state.cart.view_cart()
        st.success(result)

    elif option == "Clear Cart":
        if st.button("Clear Cart"):
            result = st.session_state.cart.clear_cart()
            st.success(result)

# Run the Streamlit app
ngrok.set_auth_token('2YcLOFA4Tjy9KncigUNeSSGKLw5_2ytuPJj7JcW6KvmDGBXkW')
#!streamlit run shopping_cart_app & npx localtunnel  port 8501
!streamlit run /content/app.py & npx localtunnel  port 8501
#!streamlit run --server.port 8501 & npx localtunnel  port 8501
````

Imagine a school management system.

**You have to design the 'Student' class using OOP concepts.The 'Student' class has the following attributes:**

a. name: Represents the name of the student.

b. age: Represents the age of the student.

c. grade: Represents the grade or class of the student.

d. student_id: Represents the unique identifier for the student.

e. attendance: Represents the attendance record of the student.

**The class should also include the following methods:**

a. update_attendance(self, date, status): Updates the attendance record of the
student for a given date with the provided status (e.g., present or absent).
b. get_attendance(self): Returns the attendance record of the student.
c. get_average_attendance(self): Calculates and returns the average
attendance percentage of the student based on their attendance record.


````python
from tabulate import tabulate
import matplotlib.pyplot as plt

class Student:
    def __init__(self, name, age, grade, student_id):
        # Attributes
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = {}

    def update_attendance(self, date, status):
        # Update attendance record for the given date
        self.attendance[date] = status
        print(f"Attendance updated for {self.name} on {date}: {status}")

    def get_attendance(self):
        # Return the attendance record
        return self.attendance

    def get_average_attendance(self):
        # Calculate and return the average attendance percentage
        total_days = len(self.attendance)
        if total_days == 0:
            return 0  # No attendance recorded yet

        present_days = sum(1 for status in self.attendance.values() if status == 'present')
        average_percentage = (present_days / total_days) * 100
        return average_percentage

    def display_attendance_table(self):
        # Display attendance record in a tabular format
        headers = ["Date", "Status"]
        rows = [(date, status) for date, status in self.attendance.items()]
        table = tabulate(rows, headers, tablefmt="grid")
        print(f"Attendance Record for {self.name}:\n{table}")

    def display_attendance_plot(self):
        # Display attendance record as a bar plot
        dates = list(self.attendance.keys())
        status = [1 if s == 'present' else 0 for s in self.attendance.values()]

        plt.figure(figsize=(10, 5))
        plt.bar(dates, status, color=['green' if s == 'present' else 'red' for s in self.attendance.values()])
        plt.title(f"Attendance Record for {self.name}")
        plt.xlabel("Date")
        plt.ylabel("Status (Present/Absent)")
        plt.show()

     # Workflow
    def student_workflow():
      # Create a student
      student1 = Student(name="John Doe", age=15, grade=10, student_id="12345")
      # Update attendance for some dates
      student1.update_attendance("2023-01-01", "present")
      student1.update_attendance("2023-01-02", "absent")
      student1.update_attendance("2023-01-03", "present")
      # Display attendance record in a tabular format
      student1.display_attendance_table()
      # Get average attendance
      average_attendance = student1.get_average_attendance()
      print(f"Average Attendance for {student1.name}: {average_attendance}%")


if __name__=="__main__":
  # Run the student workflow
  student_workflow()
  # Display attendance record as a bar plot
  student1.display_attendance_plot()
````
